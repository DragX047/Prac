```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("D:\demonic head\insurance.csv")
df.head()
```

```python
sex_replace_values={'male':0,'female':1}
region_replace_values={'southwest':0,'northwest':1,'northeast':2,'southeast':3}
smoker_replace_values={'no':0,'yes':1}
df['region']=df['region'].replace(to_replace=region_replace_values)
df['sex']=df['sex'].replace(to_replace=sex_replace_values)
df['smoker']=df['smoker'].replace(to_replace=smoker_replace_values)
df.head()
```

```python
X=df
Y=df['charges']
X_train0, X_test, Y_train0, Y_test=train_test_split(X,Y,train_size=0.82,random_state=0)
```

```python
X_train,X_val,Y_train,Y_val=train_test_split(X_train0,Y_train0,train_size=0.8,random_state=0)
input_shape=[X_train.shape[1]]
print(input_shape)
```

```python
X_train.head()
```

```python
model=keras.Sequential([
    keras.layers.Dense(128,activation=tf.nn.relu,input_shape=input_shape),
    keras.layers.Dense(128,activation=tf.nn.relu),
    keras.layers.Dense(64,activation=tf.nn.relu),
    keras.layers.Dense(1)
])
```

```python
model.compile(optimizer='adam',loss='mean_squared_error',metrics=['accuracy'])
```

```python
model.fit(X_train,Y_train,validation_data=[X_val,Y_val],epochs=400,batch_size=500,validation_batch_size=100)
```

```python
test_loss, test_acc =model.evaluate(X_test,Y_test)
```
