KNN MODEL


```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

```python
X, y = make_blobs(n_samples = 500, n_features = 2, centers = 4,cluster_std = 1.5, random_state = 4)
```

```python
#plt.style.use('seaborn')
plt.figure(figsize = (10,10),dpi=350)
plt.scatter(X[:,0], X[:,1], c=y, marker= '*',s=100,edgecolors='black')
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,train_size=0.67, random_state = 1)
```

```python
knn7 = KNeighborsClassifier(n_neighbors = 7)
knn1 = KNeighborsClassifier(n_neighbors=1)
```

```python
knn7.fit(X_train, y_train)
knn1.fit(X_train, y_train)

y_pred_7 = knn7.predict(X_test)
y_pred_1 = knn1.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy with k=5", accuracy_score(y_test, y_pred_7)*100)
print("Accuracy with k=1", accuracy_score(y_test, y_pred_1)*100)
```

```python
plt.figure(figsize = (15,5),dpi=254)
plt.subplot(1,2,1)
plt.scatter(X_test[:,0], X_test[:,1], c=y_pred_7, marker= '*', s=100,edgecolors='black')
plt.title("Predicted values with k=5", fontsize=20)

plt.subplot(1,2,2)
plt.scatter(X_test[:,0], X_test[:,1], c=y_pred_1, marker= '*', s=100,edgecolors='black')
plt.title("Predicted values with k=1", fontsize=20)
plt.show()
```
